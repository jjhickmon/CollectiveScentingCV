```python
class BeeDataset(Dataset):
    '''
    Always apply baseline transforms; augmentations optional.
    '''
    def __init__(self, root_path, baseline_transforms, augment_transforms=None, mode='train'):
        # Mode for usage of this class
        self.mode = mode

        # Paths
        self.root_path = root_path

        # Transforms
        self.baseline_transforms = baseline_transforms
        self.augment_transforms = augment_transforms
        
        # Setups
        self.store_imgs()
        self.prep_data()
        
    def store_imgs(self):
        self.img_paths = sorted(glob.glob(f"{os.path.join(self.root_path, f'cropped_imgs')}/*.png"))
        
        # Create collection of images
        self.all_imgs = []
        for path in self.img_paths:
            img = cv2.imread(path, 0)
            self.all_imgs.append(img)
            
#     def create_lookup(self):
#         # One-hot encoding vectors for each angle
#         thetas = np.arange(0, 361, 1)
#         n = len(thetas)
#         self.onehot_lookup = np.eye(n)
        
    def prep_data(self):
        self.img_dim = self.all_imgs[0].shape[:2]
        self.img_c = self.all_imgs[0][..., np.newaxis].shape[-1]
        
#         self.label_theta = int(os.path.basename(self.img_paths[0]).split("__")[1].split("_")[-1].replace('.png', ''))
#         self.head_x = float(os.path.basename(self.img_paths[0]).split("__")[2].split('_')[-1].split(',')[0])
#         self.head_y = float(os.path.basename(self.img_paths[0]).split("__")[2].split('_')[-1].split(',')[1].replace('.png', ''))
#         self.tail_x = float(os.path.basename(self.img_paths[0]).split("__")[3].split('_')[-1].split(',')[0])
#         self.tail_y = float(os.path.basename(self.img_paths[0]).split("__")[3].split('_')[-1].split(',')[1].replace('.png', ''))

    def __getitem__(self, idx):
        img = self.all_imgs[idx][..., np.newaxis]
        label_theta = int(os.path.basename(self.img_paths[idx]).split("__")[1].split("_")[-1].replace('.png', ''))
        
        # Debug
        if label_theta < 0:
            label_theta = 0
        
        head_x = float(os.path.basename(self.img_paths[idx]).split("__")[2].split('_')[-1].split(',')[0])
        head_y = float(os.path.basename(self.img_paths[idx]).split("__")[2].split('_')[-1].split(',')[1].replace('.png', ''))
        tail_x = float(os.path.basename(self.img_paths[idx]).split("__")[3].split('_')[-1].split(',')[0])
        tail_y = float(os.path.basename(self.img_paths[idx]).split("__")[3].split('_')[-1].split(',')[1].replace('.png', ''))

        # Apply optinal augmentation transforms
        if self.augment_transforms is not None:
            img = self.augment_transforms(img)
        # For all, apply baseline transforms
        if self.baseline_transforms is not None:
            img = self.baseline_transforms(img)

        return img, label_theta, head_x, head_y, tail_x, tail_y
    
    def __len__(self):
        total_num_imgs = len(self.all_imgs)
```

# Model

```python
NUM_CLASSES = 361
```

```python
def build_resnet(num_classes):
    print(f"Building resnet-18 with {num_classes} classes.")
    resnet = torchvision.models.resnet18(pretrained=True)
    num_ftrs = resnet.fc.in_features
    resnet.fc = nn.Linear(num_ftrs, num_classes)
    return resnet
```

```python
model = build_resnet(NUM_CLASSES)
```

# Check load in trained model

```python
def init_metrics():
    metrics = {
        "losses" : {
            "train" : [],
            "test"  : []
        },
        "accs" : {
            "train" : [],
            "test"  : []
        }
    }
    return metrics
```

```python
LOAD_MODEL = False
load_path = 'saved_models/ResnetOrientationModel.pt'

if LOAD_MODEL:
    print(f"Loading model from {load_path}")
    load_dict = torch.load(load_path)
    model.load_state_dict(load_dict['model'])
    optimizer.load_state_dict(load_dict['optim'])
    metrics = load_dict['metrics']
    start_epoch = load_dict['epoch']
else:
    print("Initializing new model.")
    metrics = init_metrics()
    start_epoch = 0
```

# Save model

```python
def save_model(model, optimizer, metrics, epoch_i):
    save_dict = {
        'model'   : model.state_dict(),
        'optim'   : optimizer.state_dict(),
        'metrics' : metrics,
        'epoch'   : epoch_i
    }
    torch.save(save_dict, f'../saved_models/ResnetOrientationModel.pt')
```

# Evaluation functions

```python
def get_prediction(logits):
    probs = torch.softmax(logits, dim=1)
    preds = torch.argmax(probs, dim=1)
    return preds
```

```python
def get_accuracy(y, preds):
    # For orientation, we could allow prediction to have some degree of error (5-10deg?)
    correct_sum = (preds == y).float().sum()
    # correct_sum = (abs(preds - y) <= 10).float().sum()
    accuracy = correct_sum.mul_(100.0 / len(y))
    return accuracy
```

```python
def evaluate(test_loader, model, criterion, device):
    batch_test_accs = []
    batch_test_loss = []
    with torch.no_grad():
        model.eval()
        for batch_i, (X, y, head_x, head_y, tail_x, tail_y) in enumerate(test_loader):
        # ----------------------
            X = Utils.convert_X_for_resnet(X)

            X = X.to(device)
            y = y.to(device)

            logits = model(X)
            preds = get_prediction(logits)
            accuracy = get_accuracy(y, preds)
            batch_test_accs.append(accuracy.item())

            loss = criterion(logits, y)
            batch_test_loss.append(loss.item())

            # stdout
            eval_stdout = f'\rBatch {batch_i}/{len(test_loader)} -- Accuracy: {np.mean(batch_test_accs):0.8f} -- Loss: {np.mean(batch_test_loss):0.8f}'
            sys.stdout.write(eval_stdout)
            sys.stdout.flush()

    mean_acc = sum(batch_test_accs) / len(batch_test_accs)
    mean_loss = sum(batch_test_loss) / len(batch_test_loss)
    model.train()
    return mean_acc, mean_loss
```

# Train

```python
N_EPOCHS = 5000
```

```python
%%time
try:
    model.train()
    for epoch_i in range(start_epoch, start_epoch+N_EPOCHS):
        
        # Training loop
        # ------------------------------------------------
        print("Training")
        batch_train_losses = []
        batch_train_accs = []
        for batch_i, (X_, y_, head_x, head_y, tail_x, tail_y) in enumerate(train_loader):
            X_, y_ = Utils.shuffle_batch(X_, y_)
            X_fixed = Utils.convert_X_for_resnet(X_)
            
            X_fixed = X_fixed.to(device)
            y_ = y_.to(device)
    
            # Pass through model
            logits = model(X_fixed)

            # Compute loss
            loss = criterion(logits, y_)
            batch_train_losses.append(loss.item())

            # Training step
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            # Get training accs
            with torch.no_grad():
                preds = get_prediction(logits)
                accuracy = get_accuracy(y_, preds)
            batch_train_accs.append(accuracy.item())
            
            sys.stdout.write(f'\rEpoch {epoch_i+1}/{start_epoch+N_EPOCHS} -- Batch {batch_i}/{len(train_loader)} -- Loss: {loss.item():0.8f} -- Train accuracy: {np.mean(batch_train_accs):0.4f}')
            sys.stdout.flush()
            
        metrics['losses']['train'].append((epoch_i, np.mean(batch_train_losses)))
        metrics['accs']['train'].append((epoch_i, np.mean(batch_train_accs)))
        # ------------------------------------------------

        # Test Acc
        print("\nValidation.")
        with torch.no_grad():
            test_acc, test_loss = evaluate(test_loader, model, criterion, device)
        metrics['accs']['test'].append((epoch_i, test_acc))
        metrics['losses']['test'].append((epoch_i, test_loss))
        print("\n")
        
        if epoch_i % 2 == 0:
            save_model(model, optimizer, metrics, epoch_i)

except KeyboardInterrupt:
    print("\nEnding early.")
```

# Compute accuracy with tolerance

```python
tolerance_array = np.arange(0, 40, 5)
tolerance_array
```

```python
accs_across_tolerances = []

with torch.no_grad():
    model.eval()
    for tol_i, tol in enumerate(tolerance_array):
        batch_test_accs = 0
        for batch_i, (X, y, *ignore) in enumerate(test_loader):
            # Prep data for net
            # ----------------------
            X = Utils.convert_X_for_resnet(X)
            X = X.to(device)
            y = y.to(device)
            # ----------------------

            # Run model
            logits = model(X)
            
            # Get predictions
            pred = get_prediction(logits).item()
            y = y.item()
            
            accurate = (y - tol/2) <= pred <= (y + tol/2)
            if accurate:
                batch_test_accs += 1

            # stdout
            # ----------------------------------------------------------------------
            eval_stdout = f'\rTolerance [{tol_i+1}/{len(tolerance_array)}]: {tol} -- '
            eval_stdout += f'Batch {batch_i}/{len(test_loader)} -- '
            eval_stdout += f'Accuracy: {batch_test_accs/len(test_loader)*100:0.3f}'
            sys.stdout.write(eval_stdout)
            sys.stdout.flush()
            # ----------------------------------------------------------------------

        accs_across_tolerances.append(batch_test_accs/len(test_loader)*100)
```

```python
def get_endpoint(point, angle, length):
    '''
    point - Tuple (x, y)
    angle - Angle you want your end point at in degrees.
    length - Length of the line you want to plot.
    '''
    # Unpack the first point
    x, y = point
    
    # Find the end point
    endy = y + (length * np.sin(np.radians(angle)))
    endx = x + (length * np.cos(np.radians(angle)))
    return endx, endy
```

```python
def plot_arrows(X, y, head, tail, length, ax=None, title=None):
    img = X.squeeze().permute(1,2,0).numpy()
    img = cv2.normalize(img,None,0,255,cv2.NORM_MINMAX).astype(np.uint8)
    
    midpoint_x = (float(head_x) + float(tail_x)) // 2 
    midpoint_y = (float(head_y) + float(tail_y)) // 2 

    # True line
    true_x1 = tail[0]
    true_y1 = tail[1]
    true_endx = head[0]
    true_endy = head[1]
    
    # Predicted line
    pred_endx, pred_endy = get_endpoint((midpoint_x, midpoint_y), angle=-int(preds), length=length//2)
    pred_x1, pred_y1 = get_endpoint((midpoint_x, midpoint_y), angle=-int(preds)+180, length=length//2)

    if ax:
        plot_type = ax
    else:
        plot_type = plt
        
    plot_type.annotate('', xy=(true_x1, true_y1), xytext=(true_endx, true_endy), 
               arrowprops=dict(arrowstyle="->", color='#81e35d', alpha=0.9, linewidth=2, mutation_scale=28))
    plot_type.annotate('', xy=(pred_x1, pred_y1), xytext=(pred_endx, pred_endy), 
               arrowprops=dict(arrowstyle="->", color='orange', alpha=0.9, linewidth=2, mutation_scale=28))
    
    if plot_type == ax and title:
        plot_type.set_title(f'True: {int(y)}, \n Pred: {int(preds)}')
        
    plot_type.imshow(img)
```
