# Model

```python
NUM_CLASSES = 2
```

```python
def build_resnet(num_classes):
    print(f"Building resnet-18 with {num_classes} classes.")
    resnet = torchvision.models.resnet18(pretrained=True)
    num_ftrs = resnet.fc.in_features
    resnet.fc = nn.Linear(num_ftrs, num_classes)
    return resnet
```

```python
model = build_resnet(NUM_CLASSES)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
model.to(device);
```

# Loss and optimizer

```python
LEARNING_RATE = 0.00001
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
```

# Check load in trained model

```python
def init_metrics():
    metrics = {
        "losses" : [],
        "accs" : {
            "train" : [],
            "test"  : []
        }
    }
    return metrics
```

```python
LOAD_MODEL = False
load_path = 'saved_models/ResnetScentingModel__2019-10-12__23-59-07.pt'

if LOAD_MODEL:
    print(f"Loading model from {load_path}")
    load_dict = torch.load(load_path)
    model.load_state_dict(load_dict['model'])
    optimizer.load_state_dict(load_dict['optim'])
    metrics = load_dict['metrics']
    start_epoch = load_dict['epoch']
else:
    print("Initializing new model.")
    metrics = init_metrics()
    start_epoch = 0
```

# Train on Single Batch

```python
for batch_i, (X_, y_) in enumerate(train_loader):
    break
    
X_, y_ = Utils.shuffle_batch(X_, y_)
X_fixed = Utils.convert_X_for_resnet(X_)
X_fixed = X_fixed.to(device)
y_ = y_.to(device)
```

```python
X_.shape
```

```python
N_EPOCHS = 100
```

```python
losses = []
train_accs = []
test_accs = []
```

```python
try:
    model.train()
    for epoch_i in range(start_epoch, start_epoch+N_EPOCHS):
        stdout_str = f'\rEpoch {epoch_i+1}/{start_epoch+N_EPOCHS} -- Batch {batch_i}/{len(train_loader)} -- '
        sys.stdout.write(stdout_str)
        sys.stdout.flush()

        # Pass through model
        logits = model(X_fixed)
        
        # Compute loss
        loss = criterion(logits, y_)
        losses.append(loss.item())
        
        # Training step
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # Get training accs
        with torch.no_grad():
            preds = Utils.get_prediction(logits)
            accuracy = Utils.get_accuracy(y_, preds)
        train_accs.append(accuracy)

        # Test Acc
#         print('Get test acc.')
#         with torch.no_grad():
#             test_acc = Utils.evaluate(test_loader, model, device)
#         test_accs.append(test_acc)
#         print('Done.\n')
        

except KeyboardInterrupt:
    print("\nEnding early.")
```

# Save model

```python
timestamp = datetime.strftime(datetime.now(), '%Y-%m-%d__%H-%M-%S')
```

```python
def save_model(model, optimizer, metrics, epoch_i):
    save_dict = {
        'model'   : model.state_dict(),
        'optim'   : optimizer.state_dict(),
        'metrics' : metrics,
        'epoch'   : epoch_i
    }

    torch.save(save_dict, f'saved_models/ResnetScentingModel.pt')
```

# Train 

```python
N_EPOCHS = 100
```

```python
try:
    model.train()
    for epoch_i in range(start_epoch, start_epoch+N_EPOCHS):
        for batch_i, (X_, y_) in enumerate(train_loader):
            
            X_, y_ = Utils.shuffle_batch(X_, y_)
            X_fixed = Utils.convert_X_for_resnet(X_)
            
            X_fixed = X_fixed.to(device)
            y_ = y_.to(device)
    
            # Pass through model
            logits = model(X_fixed)

            # Compute loss
            loss = criterion(logits, y_)
            metrics['losses'].append(loss.item())

            # Training step
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Get training accs
            with torch.no_grad():
                preds = Utils.get_prediction(logits)
                accuracy = Utils.get_accuracy(y_, preds)

            metrics['accs']['train'].append(accuracy)

            # Test Acc
            with torch.no_grad():
                test_acc = Utils.evaluate(test_loader, model, device)
            metrics['accs']['test'].append(test_acc)

            sys.stdout.write(f'\rEpoch {epoch_i+1}/{start_epoch+N_EPOCHS} -- Batch {batch_i}/{len(train_loader)} -- Loss: {loss.item():0.8f} -- Train accuracy: {accuracy:0.4f} -- Test accuracy: {test_acc:0.4f}')
            sys.stdout.flush()
          
        if epoch_i % 1 == 0:
            save_model(model, optimizer, metrics, epoch_i)

except KeyboardInterrupt:
    print("\nEnding early.")
```
